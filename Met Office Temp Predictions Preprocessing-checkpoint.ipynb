# Met Office Temperature Predictions using ML - Preprocessing

Known about the data:

Data collected from 36 weather stations across the UK

Columns: \
year: Year in which the measurements were taken \
month: Month in which the measurements were taken \
tmax: Mean daily maximum temperature (°C) \
tmin: Mean daily minimum temperature (°C) \
af: Days of air frost recorded that month (days) \
rain: Total rainfall (mm) \
sun: Total sunshine duration (hours) \
station: Where station measurement was recorded 

## Project outline:

- Explore and analyse sunshine hours (sun), rain in mm (rain), max temperature (tmax), min temperature (tmin), average frost days in a month (af) in London using historical data from the heathrow weather station.
- Create and analyse average temperature (tavg) in London using historical data from the heathrow weather station.
- Use various machine learning models to predcit the mean temp, min temp and max temp in London using historical data from the heathrow weather station (1950-2020).

# Drop unwanted columns and rows

```python
# Make a copy of the df 
weather_data_sorted = weather_data.copy()

# Sort df by years
weather_data_sorted = weather_data_sorted.sort_values('year')

# View
weather_data_sorted
```

```python
# Drop af and sun columns as wnat to retain as many historical years of data 
# and these 2 columns have a lot of missing data for entire years
# so imputation would be synthetic and may make the model less accurate.

weather_data_sorted.drop(columns=['sun', 'af'], inplace=True)
weather_data_sorted
```

```python
# Drop rows with years from 1853 up to 1948 
# Heathrow only ahs data starting from 1948 by the looks of it

# Drop specified row range from df
weather_data_sorted = weather_data_sorted[weather_data_sorted['year'] >= 1948]

# Reset index
weather_data_sorted = weather_data_sorted.reset_index(drop=True)

# Drop all rows with station names other than heathrow as only looking at London for now
weather_data_sorted = weather_data_sorted[weather_data_sorted['station'] == 'heathrow']

# Sort data by years column
weather_data_sorted = weather_data_sorted.sort_values(by=['year', 'month'])

# Reset index
weather_data_sorted.reset_index(drop=True, inplace=True)

# View 
weather_data_sorted

# Now 870 rows and 5 columns, small data set for ml project
```

```python
# Change year and month from float to integers 
weather_data_sorted['year'] = weather_data_sorted['year'].astype(int)
weather_data_sorted['month'] = weather_data_sorted['month'].astype(int)

# View
weather_data_sorted
```

```python
# Sort and view all unique years in df
sorted_unique_years = weather_data_sorted['year'].unique()

# Number of unique years 
num_unique_years = len(sorted_unique_years)

# Print
print('Number of unique years:', num_unique_years)
print('Unique years:', sorted_unique_years)

# Can now see that we have 73 years of data from 1948 to 2020 (2020 not complete year)
```

### Create average temperature variable

```python
# Calculate the average temp by summing tmin and tmax and dividing by 2
# assigning output to a new column called tmean and concat together into weather_df with other variables
weather_data_sorted['tmean'] = (weather_data_sorted['tmin'] + weather_data_sorted['tmax']) / 2

# Concat tmean in a new df called weather_df with year, month, tmin and tmax.
weather_df = pd.concat([weather_data_sorted[['year', 'month', 'tmin', 'tmax', 'rain']],
                        weather_data_sorted['tmean']], axis=1)

# View
weather_df
```

### Correlation matrix 

```python
# Created a correlation heatmap to view correlations between variables
corrMatrix = weather_df.corr()
sns.heatmap(corrMatrix, annot = True)
plt.show()

# Ignoring year and month. As predicted temp varibles are highly positively correlated with each other.
# Rain is weakly positively correlated with tmin and weakly negatively correlated with tmax and tmean.
```
